# Adverserial search

The Tic-Tac-Toe game starts on a 3x3 grid with two players "X" and "O" who take turns and play. The rules are as follows: each player gets a turn with player "X" (resp. "O") writing an "X" (resp. "O") in an empty cell of the grid. The game starts with the move of the "O" player. The first player to write on three horizontal or vertical or diagonal cells wins.

(a) Use the minimax strategy to design an AI that plays the game optimally. The leaf nodes where "X" wins gets 1 and "O" wins gets -1 and neither wins gets a zero. A sample game play is given below.

```python
import random
import copy
global state                          # declare global variable
# function to check if there are any moves left on the board
def movesLeft(state):
    for i in range(1,10,1):
        if state[i]=="":              #condition to check if there are any empty spaces
            return 1
    return 0
# function to return a list of empty positions on the board
def pos_empty(state):
    l=[]                              #declaring the list to store the empty positions
    for i in range(1,10,1):
        if state[i]=="":
            l.append(i)               #appending the empty positions
    return l
# function to evaluate the score of the board
def score(state,player_X,player_O):
    c = []
    # check for winning condition in rows
    for i in range(1,8,3):
        if(state[i]==state[i+1] and state[i+1]==state[i+2] and state[i+2]==state[i]):
            if(state[i]==player_X):   # check if Player X wins
                return 1
            elif(state[i]==player_O): # check if Player O wins
                return -1
    # check for winning condition in columns
    for i in range(1,4,1):
        if(state[i]==state[i+3] and state[i+3]==state[i+6] and state[i+6]==state[i]):
            if(state[i]==player_X):   # check if Player X wins
                return 1
            elif(state[i]==player_O): # check if Player O wins
                return -1
    # check for winning condition in diagonal
    if(state[1]==state[5] and state[5]==state[9] and state[9]==state[1]):
        if(state[1]==player_X):       # check if Player X wins
            return 1
        elif(state[1]==player_O):     # check if Player O wins
            return -1
    if(state[3]==state[5] and state[5]==state[7] and state[7]==state[3]):
        if(state[3]==player_X):
            return 1
        elif(state[3]==player_O):
            return -1
    #return 0 if game is still going on or game ends in DRAW
    return 0
# function to return a list of all possible next states of a given state
def neighbour_state(state,a):
    neighbour=[]                            #declaring list to store the child of given state
    emp = pos_empty(state)                  #storing the empty positions 
    for i in emp:
        child = copy.deepcopy(state)
        child[i] = a
        neighbour.append(child)             #appending the childs in the list
    return neighbour
# function to return the maximum score from a given state for the AI player X
def maxValue(g_state):
    if(movesLeft(g_state)==0 or score(g_state,"X","O")!=0):    #checking if the state is a termination state
        return score(g_state,"X","O")
    else:
        next_state=neighbour_state(g_state,"X")
        optimal=[]                                       #declaring the list to store the utility 
        for i in next_state:
            optimal.append(minValue(i))                  #appending the utilities of all childs
        return max(optimal)                              #returning the maximum among the utilities of all childs
# function to return the minimum score from a given state for the player O
def minValue(g_state):
    if(movesLeft(g_state)==0 or score(g_state,"X","O")!=0):     #checking if the state is a termination state
        return score(g_state,"X","O")
    else:
        next_state=neighbour_state(g_state,"O")
        optimal=[]                                              #declaring the list to store the utility
        for i in next_state:
            optimal.append(maxValue(i))                         #appending the utilities of all childs
        return min(optimal)                                     #returning the minimum among the utilities of all childs
#function to return optimal move by the AI after the turn of player O
def minimax():
    global state
    AImove = neighbour_state(state,"X")
    optimal=[]                                #declaring the list to store the utility        
    for i in AImove:
        optimal.append(minValue(i))           #appending the utilities of all childs of given state
    m = max(optimal)                          # storing the maximum utility among the utilities of all childs
    lst=[]
    for i in range(len(optimal)):
        if(optimal[i]==m):
            lst.append(i)                     # appending the childs with maximum utility
    randomstate = random.choice(lst)          #choosing the next move randomly from the list lst
    state = AImove[randomstate]
    print_board(state)
    print("\n")
    return state                              #returning the state after the move of AI
#function to print the board for the given state
def print_board(state):
    for i in range(1,10,1):
        if(i%3==0):
            if(i==9):
                print(state[i])
            else:
                print(state[i])
                print("-+-+-")
        else:
            if(state[i]=="" and i!=9):
                print(state[i],end=" |")
            elif(i!=9):
                print(state[i],end="|")
#functioon to check if the position entered by the player X is valid or not
def invalid(state,pos):
    if(state[pos]!=""):
        return 1
    return 0
# function to play the game
def play(state):
    while score(state,"X","O")==0 and movesLeft(state)!=0:
        p = int(input(f"turn of player X , enter the position "))       #taking the input of position   
        while(invalid(state,p)==1):                                     #checking for invalid input
            p = int(input(f"Invalid move \nEnter the position again ")) #taking the input again
        state[p]="O"                                                    #updating the state
        print_board(state)                                              #printing the state
        print("\n")
        if(score(state,"X","O")!=0 or movesLeft(state)==0):             #checking for the termination condition
            break
        else:
            state=minimax()                                             #calling the minimax function
state={}                                    #initialising the state of the game
for i in range(1,10,1):
    state[i]=""
print_board(state)
print("\n")
play(state)                                 #calling the play function to start the game
#checking and printing the winning condition of Player X or Player O or DRAW
if(score(state,"X","O")==0):
    print("GAME DRAW")
elif(score(state,"X","O")==1):
    print("Player X won")
else:
    print("Player O won")
```
